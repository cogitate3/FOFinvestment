```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
```

```python
df0 = pd.read_excel(r"d:\数据-需备份\百度云\余铮研究\新研究201905起\FOF投资\公募fof基金基本数据-20210302.xlsx")
df0.columns 
```

```python
# 修改导出时凌乱的列标签
df0.columns = ['证券代码', '证券简称', '基金类型', '单位净值元',
       '基金份额份', '投资风格', '基金管理人', '基金托管人',
       '管理费率%', '托管费率%', '基金经理(现任)', '基金存续期', '基金成立日',
       '基金到期日']
df0.columns
```

```python
# 把份额单位改为亿份
df0["基金份额份"] = df0["基金份额份"]/(10**8)
df0.rename(columns={'基金份额份': '基金份额亿份'}, inplace=True) 
df0.基金份额亿份
```

```python
# 去掉wind导出数据时留的尾巴
lastrow = len(df0.index) - 1
df1 = df0.drop(labels=[lastrow-1,lastrow],axis=0)
```

```python
# 分组统计结果按照份额排序
result2 = df1.groupby(['基金管理人']).sum().sort_values(by=["基金份额亿份"],ascending=False)
result2
```

```python
# 保存排序结果
# result2.to_excel("公募fof基金基本数据-20210302-按公司分类求和.xlsx")
```

```python
# 选出基金份额大于20亿份的各个管理人的各个基金
over20 = df1.loc[df1['基金管理人'].isin(result2.loc[(result2['基金份额亿份'] >= 20)].index.tolist())].sort_values(by=["基金管理人","基金份额亿份"],ascending=False)
over20.to_excel("公募fof基金基本数据-20210302-按公司分类求和over20.xlsx")
```

```python
# 选出基金份额大于10亿份的各个管理人的各个基金
top10 = df1.loc[df1['基金管理人'].isin(result2.loc[(result2['基金份额亿份'] >= 10)].index.tolist())].sort_values(by=["基金管理人","基金份额亿份"],ascending=False)
top10
top10.to_excel("公募fof基金基本数据-20210302-按公司分类求和top10.xlsx")
```
